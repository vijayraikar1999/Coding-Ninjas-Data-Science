```python
import numpy as np
```

```python
x = np.arange(0, 5, 0.1)
y = x**3

plt.plot(x, y)
plt.show()
```

```python
a = [3, 4, 5]
plt.plot(a, marker = '*')
plt.show()
```

It's gonna take the <b>a</b> as y axis points and <b>range(n)</b> as x axis points where <b>n</b> is the size of <b>a</b>.

```python
x = np.arange(0, 5, 0.1)
y = x**3

plt.plot(x, y, color = 'green', marker = '*')
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3

plt.plot(x, y, color = 'black', linewidth = 2)
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3

plt.ylabel('X^3')
plt.xlabel('X')
plt.title('Matplotlib demo')

plt.plot(x, y, color = 'black', linewidth = 2)
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3
y2 = x**2

plt.plot(x, y, color = 'black', linewidth = 2)

plt.plot(x, y2, color = 'green', linewidth = 2)

plt.ylabel('y')
plt.xlabel('x')
plt.title('Matplotlib demo')
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3
y2 = x**2

plt.plot(x, y, color = 'black', linewidth = 3, label = 'x^3')

plt.plot(x, y2, color = 'green', linewidth = 3, label = 'x^2')

plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title('Matplotlib demo')
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**3
y2 = x**2

plt.plot(x, y, color = 'black', linewidth = 3, label = 'x^3')

plt.plot(x, y2, color = 'green', linewidth = 3, label = 'x^2')

plt.axis([0, 10, 0, 100])

plt.ylabel('y')
plt.xlabel('x')

plt.grid()

plt.legend()
plt.title('Matplotlib demo')
plt.show()
```

### Bubble Charts

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

```python
plt.scatter(year, salary, s = population, c = 'g')
plt.title('Bubble Chart', fontsize = 18)
plt.xlabel('year', fontsize = 15)
plt.ylabel('salary', fontsize = 15)
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = 'g', alpha = 0.3)
plt.title('Bubble Chart', fontsize = 18)
plt.xlabel('year', fontsize = 15)
plt.ylabel('salary', fontsize = 15)
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = ['r', 'g', 'b', 'cyan', 'grey', 'hotpink'])
plt.title('Bubble Chart', fontsize = 18)
plt.xlabel('year', fontsize = 15)
plt.ylabel('salary', fontsize = 15)
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = np.arange(len(year)))
plt.title('Bubble Chart', fontsize = 18)
plt.xlabel('year', fontsize = 15)
plt.ylabel('salary', fontsize = 15)
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = np.arange(len(year)), edgecolor = 'b')
plt.title('Bubble Chart', fontsize = 18)
plt.xlabel('year', fontsize = 15)
plt.ylabel('salary', fontsize = 15)
plt.show()
```

### Pie Charts

```python
sizes = [3, 4, 6, 2]
plt.pie(sizes)
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors = ['cyan', 'purple', 'hotpink', 'teal']
labels = ['A', 'B', 'C', 'D']
plt.title('Split among classes')
plt.pie(sizes, colors = colors, labels = labels)
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors = ['cyan', 'purple', 'hotpink', 'teal']
labels = ['A', 'B', 'C', 'D']
plt.title('Split among classes')
plt.pie(sizes, colors = colors, labels = labels, autopct = '%.2f%%')
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors = ['cyan', 'purple', 'hotpink', 'teal']
explode = [0.1, 0.2, 0, 0]
labels = ['A', 'B', 'C', 'D']
plt.title('Split among classes')
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct = '%.2f%%')
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors = ['cyan', 'purple', 'hotpink', 'teal']
explode = [0.1, 0.2, 0, 0]
labels = ['A', 'B', 'C', 'D']
plt.title('Split among classes')
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct = '%.2f%%', counterclock = False)
plt.show()
```

```python
sizes = [3, 4, 6, 2]
colors = ['cyan', 'purple', 'hotpink', 'teal']
explode = [0.1, 0.2, 0, 0]
labels = ['A', 'B', 'C', 'D']
plt.title('Split among classes')
plt.pie(sizes, colors = colors, explode = explode, labels = labels, autopct = '%.2f%%', counterclock = False, startangle = 100)
plt.show()
```
