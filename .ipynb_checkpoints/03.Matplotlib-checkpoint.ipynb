```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x2 = [3, 4, 5]
y2 = [2, 4, 6]

plt.scatter(x, y)
plt.plot(x, y)

plt.scatter(x2, y2)
plt.plot(x2, y2)

plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y)

plt.scatter(x2, y2, marker = '*')
plt.plot(x2, y2)

plt.show()
```

```python
plt.plot(x, y, 'g')
plt.show()
```
